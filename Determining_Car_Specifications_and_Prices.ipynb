## Data Preparation

```python
# viewing data with null values in the Engine_size feature
df[df['Engine_size'].isna() == True]
```

```python
# deleting records with a null value in the Engine_size column
df = df.drop(df[df['Engine_size'].isna() == True].index)
df.isna().sum()
```

```python
# viewing data with null values in the Price_in_thousands feature
df[df['Price_in_thousands'].isna() == True]
```

```python
# imputation of values using the mean for the Price_in_thousands feature
df['Price_in_thousands'] = df['Price_in_thousands'].fillna(df['Price_in_thousands'].mean())
df.isna().sum()
```

```python
# viewing data with null values in the Curb_weight feature
df[df['Curb_weight'].isna() == True]
```

```python
# imputation of values using the mean for the Curb_weight feature
df['Curb_weight'] = df['Curb_weight'].fillna(df['Curb_weight'].mean())
df.isna().sum()
```

```python
# viewing data with null values in the Fuel_efficiency feature
df[df['Fuel_efficiency'].isna() == True]
```

```python
# imputation of values using the mean for the Fuel_efficiency feature
df['Fuel_efficiency'] = df['Fuel_efficiency'].fillna(df['Fuel_efficiency'].mean())
df.isna().sum()
```

```python
# viewing data with null values in the Curb_weight feature
df[df['Power_perf_factor'].isna() == True]
```

```python
# imputation of values using the mean for the Power_perf_factor feature
df['Power_perf_factor'] = df['Power_perf_factor'].fillna(df['Power_perf_factor'].mean())
df.isna().sum()
```

```python
# viewing data with null values in the __year_resale_value feature
df[df['__year_resale_value'].isna() == True]
```

```python
# the difference between the new price and the used price
df[['Price_in_thousands', '__year_resale_value']]
```

```python
# create a variable that stores the average difference between the prices of new and used cars
reduction_average = pd.Series(df['Price_in_thousands'] - df['__year_resale_value']).mean()
reduction_average
```

```python
# imputing values for the __year_resale_value feature
df['__year_resale_value'] = df['__year_resale_value'].fillna(df['Price_in_thousands'].sub(reduction_average))
df.isna().sum()
```

## Modelling

#### Import Library

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, r2_score
```

#### Feature Selection

```python
pred_data = df_to_analyze.drop(['Model', 'Manufacturer', '__year_resale_value', 'Sales_in_thousands', 'Vehicle_type'], axis=1)
pred_data
```

```python
x = pred_data.drop('Price_in_thousands', axis=1)
y = pred_data['Price_in_thousands']
```

```python
x.shape
```

#### Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

#### Create a regression model

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

#### Conduct a performance evaluation

```python
y_pred = model.predict(x_test)
y_pred
```

```python
rmse = root_mean_squared_error(y_test, y_pred)
r2score = r2_score(y_test, y_pred)
print('RMSE = ', rmse)
print('R2 Score = ', r2score)
```

```python
results_df = pd.DataFrame({'Actual Sales': y_test, 'Predicted Sales': y_pred})
results_df
```

```python
fig = px.scatter(results_df, x='Predicted Sales', y='Actual Sales', trendline='ols')
fig.update_layout(
    title='Actual vs Predicted Sales',
    xaxis_title='Predicted Sales',
    yaxis_title='Actual Sales'
)
fig.show()
```
